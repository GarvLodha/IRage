# Replace '99999999'or '-99999999' with mean

```python
#inf = 9e999

def replace(var):
    if (var == 99999999) or (var == -99999999):
        #return ''
        #return data[var].mean
        return float('NaN')
        #return inf - inf
    else:
        return var
        
```

```python
data['2nd Best Buy Price Modified'] = data['2nd Best Buy Price'].apply(replace)
#data['2nd Best Buy Price'] = data['2nd Best Buy Price'].apply(replace)
sum(data['2nd Best Buy Price Modified'].isnull())
#sum(data['2nd Best Buy Price'].isnull())
```

```python
data['2nd Best Buy Price Modified'] = data['2nd Best Buy Price'].apply(lambda x: data['2nd Best Buy Price'].mean() if ((x == -99999999) or (x == 99999999)) else x)
#data['2nd Best Buy Price'] = data['2nd Best Buy Price'].apply(lambda x: data['2nd Best Buy Price'].mean() if ((x == -99999999) or (x == 99999999)) else x)
```
